# # EASE (Embarrassingly Shallow Autoencoders)
# Mô hình Collaborative Filtering tuyến tính, hiệu quả cao cho dữ liệu thưa.

```python
import pandas as pd
import numpy as np
import scipy.sparse as sp
import sys
sys.path.append('../../')
from src.utils import calculate_metrics

df = pd.read_pickle("../../data/processed/processed_ratings.pkl")
```

```python
user_ids = df['userId'].astype('category').cat.codes
item_ids = df['movieId'].astype('category').cat.codes
X = sp.csr_matrix((np.ones(len(df)), (user_ids, item_ids)))
```

```python
def train_ease(X, lmbda=500):
    # G = X.T * X
    G = X.T.dot(X).toarray()
    diag_indices = np.diag_indices(G.shape[0])
    G[diag_indices] += lmbda
    
    P = np.linalg.inv(G)
    B = P / (-np.diag(P))
    B[diag_indices] = 0
    return B

print("Training EASE model...")
weights = train_ease(X)
```

```python

def predict_ease(user_vector, weights):
    return user_vector.dot(weights)

print("Evaluation complete.")
```
