# # BERT-NeuMF Hybrid Model
# Kết hợp kiến trúc Neural Matrix Factorization với thông tin ngữ nghĩa từ PhoBERT.

```python
import torch
import torch.nn as nn
import sys
sys.path.append('../../')
from src.utils import calculate_metrics
```

```python
class BERT_NeuMF(nn.Module):
    def __init__(self, n_users, n_items, bert_dim=768, factor=64):
        super().__init__()
        self.user_embed = nn.Embedding(n_users, factor)
        self.item_embed = nn.Embedding(n_items, factor)
        self.content_dense = nn.Linear(bert_dim, factor)
        
        self.fc_layers = nn.Sequential(
            nn.Linear(factor * 3, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 1)
        )

    def forward(self, u, i, b):
        combined = torch.cat([self.user_embed(u), self.item_embed(i), self.content_dense(b)], dim=-1)
        return torch.sigmoid(self.fc_layers(combined))
```
